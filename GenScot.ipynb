```python
# @title Default title text
from matplotlib import pyplot as plt
import pandas as pd
sex_data = pd.DataFrame({'sex': ['F', 'M'], 'Frequency.2': [14151, 9925]})
plt.figure(figsize=(6, 4))
plt.bar(sex_data['sex'], sex_data['Frequency.2'], color=['orchid','skyblue',])
plt.xlabel('Sex')
plt.title('Sex Distribution')
plt.ylabel('Number of Participants')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# @title Default title text
plt.figure(figsize=(6, 4))
plt.pie(sex_data['Frequency.2'], labels=sex_data['sex'], autopct='%1.1f%%', colors=['orchid', 'skyblue'])
plt.title('Sex Distribution of Participants')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
areas = ["Tayside", "Glasgow", "Perthshire", "Grampian", "Ayrshire"]
frequencies = [10896, 9415, 2488, 1196, 72]

# Plot
plt.figure(figsize=(6,4.5))
plt.bar(areas, frequencies, color="skyblue", )

# Labels and title
plt.title("Area Frequencies")
plt.xlabel("Areas")
plt.ylabel("Frequency")

# Show values on bars
for i, freq in enumerate(frequencies):
    plt.text(i, freq + 100, str(freq), ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
